Now that each member of the team has a local copy of the 'to-do-list' project, they can start working on their tasks. Wendy, the programmer, added a skeleton HTML page called 'index.html' to her Working Directory. Which command should she run to add the HTML file to her Staging Area?

##### Answer:
A. git clone <br>
B. git init<br>
C. git pull<br>
D. git push<br>
E. git commit<br>
F. git add <br>

Well done, the 'git add .' command is the correct choice for this operation. You can now see that the 'index.html' file has been added to her Staging Area.



Now Wendy has successfully added the ‘index.html’ file to her Staging Area. Which command should she run to add the ‘index.html’ file to her Local Repository?

Well done, the 'git commit -m' command is the correct choice for this operation. You can now see that the 'index.html' file has been added to her Local Repository.



Now Wendy is ready to add her changes to the GitHub repository. Which command should she run to add the ‘index.html’ file to the Remote Repository?


Well done, the 'git push' command is the correct choice for this operation. You can now see that the 'index.html' file has been added to the Remote Repository.
